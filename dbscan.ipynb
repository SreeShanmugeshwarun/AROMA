```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

```

```python
# Data points as (x, y)
points = np.array([
    [3, 7],   # P1
    [4, 6],   # P2
    [5, 5],   # P3
    [6, 4],   # P4
    [7, 3],   # P5
    [6, 2],   # P6
    [7, 2],   # P7
    [8, 4],   # P8
    [3, 3],   # P9
    [2, 6],   # P10
    [3, 5],   # P11
    [2, 4]    # P12
])

# Apply DBSCAN
db = DBSCAN(eps=1.5, min_samples=2)
labels = db.fit_predict(points)

# Plotting
plt.figure(figsize=(8, 6))
unique_labels = set(labels)

colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']
for label in unique_labels:
    cluster_points = points[labels == label]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {label}' if label != -1 else 'Noise', s=100)

# Annotate the points with P1, P2...
for i, (x, y) in enumerate(points):
    plt.text(x + 0.1, y + 0.1, f'P{i+1}', fontsize=9)

plt.title('DBSCAN Clustering on Custom Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.show()
```
